## Conclusion

This exploratory data analysis of Delhi house prices reveals several key factors influencing property values in the dataset:

1.  **Data Quality:** Initial inspection showed missing values in 'Bathroom', 'Parking', 'Type', and 'Per_Sqft'. These were imputed using appropriate central tendency measures (mode for 'Type', median for the others) to allow for further analysis.

2.  **Key Numerical Drivers:** Correlation analysis identified 'Bathroom', 'Area', and 'BHK' as the numerical features with the strongest positive correlation with 'Price'. Visualizations supported this, showing a general trend of increasing price with higher values in these features.

3.  **Categorical Feature Impact:** The pie charts demonstrated how the total value of properties is distributed based on 'Furnishing' status, 'Status' (Ready to Move vs Almost Ready), and 'Transaction' type (Resale vs New Property). Each category holds a significant portion of the total market value represented in the dataset.

4.  **Geographical Variation:** Grouping by 'District' and visualizing average prices confirmed that property values vary significantly across different regions of Delhi. The scatter plot colored by 'Type' also highlighted price variations across numerous specific 'Location's.

In summary, while numerical features like Area, BHK, and Bathroom count are important predictors, geographical location (District and specific Locality) and property type also play a significant role in determining house prices in this Delhi dataset.
